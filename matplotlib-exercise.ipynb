```python
hd = pd.read_csv('heart-disease.csv')
hd
```

```python
plt.style.use('default')

# Create the figure and its axes with the size set
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(7.5,10), sharex=True)

# create a scatter plot with the variable of 'Age' and 'Cholestrol' and showing whether they had heart disease with color
scatter = ax0.scatter(x=hd['age'], y=hd['chol'], c=hd['target'])

#setting the labels and titles of the axis
ax0.set(title='Heart Disease & Cholestrol Levels', ylabel='Cholestrol')

# creating a legend to clarify the colors, notice when the plot was created it was set to the variable 'scatter'
ax0.legend(*scatter.legend_elements(), title='Target')

# creating a horizontal line that shows the mean value of the cholestrol level
ax0.axhline(y=hd['chol'].mean(), linestyle='--');


### ----- second axis ----- ###

scatter2 = ax1.scatter(x=hd['age'], y=hd['thalach'], c=hd['target'])

ax1.set(title='Levels of Thalach and Relevance to Heart Disease',
       xlabel='Age',
       ylabel='Thalach Levels')
ax1.legend(*scatter2.legend_elements(), title='Target')

ax1.axhline(y=hd['thalach'].mean(), linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16);
```
